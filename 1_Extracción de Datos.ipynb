**CARGA DE ARCHIVO XLSX**

```python
#Importamos la librería requerida para poder cargar el archivo en un DataFrame para comenzar a trabajar con la base de datos.
import pandas as pd
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

**FILTROS BASICOS PARA VARIABLES CUALITATIVAS**

```python
#Filtro por objeto: Muestra únicamente las filas que corresponden al cliente llamado "ELICAMEX".
# Uso: Identificar la información de un solo cliente
filtro1=df[df["NOMBRE_CLIENTE"] == "ELICAMEX"]
filtro1
```

```python
#Filtro por filas: Muestra filas consecutivas
#Nos permite conocer la información capturada de la celda 200-210
filtro2= df.iloc[200:211,: ]         
filtro2
```

```python
#Filtro por columnas: Muestra columnas no consecutivas 
#Nos permite conocer la información de solo las columnas: NOMBRE_VENDEDOR, CANT, DESCR Y MARGEN_TOTAL_CALCULADO
filtro3=df.iloc[ : , [2,4,6,13]]
filtro3
```

```python
#Filtro de filas vs columnas
#Paso 1: Leemos el contenido del archivo
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=6)
df1
#Paso 2: Aplicamos Filtro de filas vs columnas 
#Nos permite conocer el margen total calculado para productos específicos
filtro4=df1.loc[["PRE-SEPARADOR CINTA 7.8 OHMSX 3MT LONG","TEA K ELEMENTO 12CMX EXT.40CM CERAMICO","BAN MICA 120mmØx50mm 850W 440V TTORNILLO"], ["MARGEN_TOTAL_CALCULADO"]]
filtro4
```

```python
#Filtro por cabecera: Muestra la infromación de los primeros 15 registros.
filtro5=df.head(16)
filtro5
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion: Extrae todas las filas del DataFrame donde la transacciones o partidas con subtotales mayores a los 50,000
filtro6=df[df["SUBTOTAL_PARTIDA"] > 50000]
filtro6
```

**FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y: Extrae todas las filas donde la cantidad es mayor a 30 y el nombre del vendedor es "ALFREDO CANELA"
filtro7 = df[(df["CANT"] > 30) & (df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
filtro7
```

```python
#Filtro O:  Extrae las filas con precios unitarios altos (20,000 o más) o filas que tienen el código de artículo "TUE00207".
filtro8 = df[(df["PRECIO_UNITARIO"] >= 20000) | (df["CVE_ART"] == "TUE00207")]
filtro8
```

```python
# Filtro NOT:  Selecciona las filas donde el costo unitario no es 0 y la descripción no es "PT-100 4MMØX10CM EXT 5MT TEFLON 2HILOS".
filtro9 = df[~(df["COSTO_UNITARIO"] == "0") & ~(df["DESCR"] == "PT-100 4MMØX10CM EXT 5MT TEFLON 2HILOS")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Filtro por objeto.csv")
filtro2.to_csv("Filtro por filas.csv")
filtro3.to_csv("Filtro por columnas.csv")
filtro4.to_csv("Filtro de filas vs columnas.csv")
filtro5.to_csv("Filtro por cabecera.csv")
filtro6.to_csv("Filtro por comparacion.csv")
filtro7.to_csv("Filtro Y.csv")
filtro8.to_csv("Filtro O.csv")
filtro9.to_csv("Filtro NOT.csv")
```
